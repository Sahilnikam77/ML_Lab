```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("Dataset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Time_of_Day"] = le.fit_transform(df["Time_of_Day"])
```

```python
correlation_matrix = df.corr()
```

```python
print(correlation_matrix)
```

```python
plt.figure(figsize=(10,6))  # Adjust figure size
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("Original DataFrame:\n", df)
print("\nDataFrame after removing outliers:\n", df_no_outliers)
```

```python
import matplotlib.pyplot as plt

# Plot the original data
plt.figure(figsize=(10, 5))
plt.plot(df.index, df.mean(axis=1), label="Original Data", linestyle='dashed', color='red')

# Plot the cleaned data (without outliers)
plt.plot(df_no_outliers.index, df_no_outliers.mean(axis=1), label="Without Outliers", color='blue')

# Labels and title
plt.xlabel("Index")
plt.ylabel("Mean Value of Rows")
plt.title("Line Graph Before and After Removing Outliers")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
x=df.drop(columns=['AC_Temp'])
```

```python
y=df['AC_Temp']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=20,random_state=40)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
print(mse)
```

```python
r2=r2_score(y_test,y_pred)
```

```python
print(r2)
```
